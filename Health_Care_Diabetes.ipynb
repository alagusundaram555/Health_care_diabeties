```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state =10)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
knn_model = KNeighborsClassifier(n_neighbors=10)
knn_model.fit(X_train, Y_train)
Y_pred = knn_model.predict(X_test)
knn_accuracy = round(accuracy_score(Y_pred, Y_test), 2)*100
knn_accuracy
```

```python
KFold_validation = KFold(8)
model_out = cross_val_score(knn_model,X,Y,cv = KFold_validation)
print(model_out)
print(np.mean(model_out))
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
SKfold = StratifiedKFold(n_splits = 8)
SKmodel_out = cross_val_score(knn_model,X,Y,cv = SKfold)
print(SKmodel_out)
print(np.mean(SKmodel_out))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
con_matrix = confusion_matrix(Y,knn_model.predict(X))
con_matrix
```

```python
print(classification_report(Y,knn_model.predict(X)))
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
knn_model_ROC = KNeighborsClassifier(n_neighbors=4)
knn_model_ROC.fit(X_train, Y_train)
knn_ROC_out = knn_model_ROC.predict_proba(X)

# Sklearn has a very potent method roc_curve() which computes the ROC for your classifier in a matter of seconds! It returns the FPR, TPR, and threshold values: calculate roc curve
fpr1, tpr1, thresholds1 = roc_curve(Y, knn_ROC_out[:, 1],pos_label=1)

# roc curve for tpr = fpr 
random_probs = [0 for i in range(len(Y))]
p_fpr, p_tpr, _ = roc_curve(Y, random_probs, pos_label=1)

# plot no skill
plt.plot(p_fpr, p_tpr, linestyle='--',color='blue')
plt.plot(fpr1, tpr1, linestyle='--',color='green', label='KNN')

# plot the roc curve for the model
plt.title('ROC curve')
# x label
plt.xlabel('False Positive Rate')
# y label
plt.ylabel('True Positive rate')
#plt.plot(fpr, tpr, marker='.')
plt.legend(loc='best')
plt.show();
# keep probabilities for the positive outcome only
#The AUC score can be computed using the roc_auc_score() method of sklearn: calculate AUC
auc_KNN = roc_auc_score(Y, knn_ROC_out[:, 1])
print('AUC KNN: %.5f' % auc_KNN)
```
